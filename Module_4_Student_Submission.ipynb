## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

#created a new column called subtotal
client_df['subtotal'] = client_df['unit_price'] * client_df['qty']

#created a new dataframe with the columns client_id, unit_price, qty, and subtotal
subtotal_df = client_df[['client_id','unit_price' , 'qty', 'subtotal']]

#sorted the subtotal column in descending order
subtotal_sorted_df = subtotal_df.sort_values(by='subtotal', ascending=False)

#converted the subtotal column to an integer
subtotal_sorted_df['subtotal'] = subtotal_sorted_df['subtotal'].astype(int)

#displayed the top 5 rows of the dataframe
subtotal_sorted_df.head()

```

```python
# Create column that calculates the total weight of order
client_df['total_weight'] = client_df['unit_weight'] * client_df['qty']

#Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# Define a function
def shipping_price(total_weight):
    if total_weight > 50:
        return total_weight * 7
    else:
        return total_weight * 10
    


# Apply the function to shipping_price column
client_df['shipping_price'] = client_df['total_weight'].apply(shipping_price)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
client_df['total_price'] = (client_df['subtotal'] + client_df['shipping_price']) * 1.0925


client_df[['client_id', 'subtotal', 'shipping_price', 'total_price']].head()

```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

client_df['line_cost'] = client_df['unit_cost'] * client_df['qty'] + client_df['shipping_price']
client_df[['client_id', 'unit_cost', 'qty', 'shipping_price', 'line_cost', 'subtotal']].head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
client_df['line_profit'] = client_df['subtotal'] - client_df['line_cost']
client_df[['client_id', 'line_cost', 'subtotal', 'line_profit']].head()
```

Use the new columns with confirmed values to find the following information.

```python
# Create a summary DataFrame showing the totals for the
# top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

summary_df = client_df[['client_id', 'qty', 'shipping_price', 'subtotal', 'line_profit']].groupby('client_id').sum()
summary_df = summary_df.sort_values(by='line_profit', ascending=False).head(5)
summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.


summary_df['qty'] = summary_df['qty'].astype(int)
summary_df['Shipping Costs'] = summary_df['shipping_price'].astype(int) / 1000000
summary_df['Total'] = summary_df['subtotal'].astype(int) / 1000000
summary_df['Profit'] = summary_df['line_profit'].astype(int) / 1000000
summary_df = summary_df[['qty', 'Shipping Costs', 'Total', 'Profit']]
summary_df.head()
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summary_df = summary_df.sort_values(by='Profit', ascending=False)
summary_df.head()
```
